```python
import pandas as pd
from pandas import read_csv
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from xgboost import plot_importance
from matplotlib import pyplot
```

```python
df = read_csv('cancer_patient_data_sets.csv', delimiter=",")
df.info()
```

```python
# Mapping values to be ML friendly
df["level"] = df["level"].map({"Low": 1,"Medium": 2,"High": 3})
df.head(5)
```

```python
# Extract only numerical part for each row of the column and convert "patient_id" to int64
df['patient_id'] = df['patient_id'].astype(str).str.extract('(\d+)', expand=False)
df = df.apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
# Save CSV
df.to_csv('Lung Cancer Detection.csv', encoding='utf-8', index=False)
```

```python
# Load new CSV
df = loadtxt('Lung Cancer Detection.csv', delimiter=",",skiprows=1)
df
```

```python


X = df[:,0:24]
y = df[:,24]
# fit model on training data
model = XGBClassifier()
model.fit(X, y)
# plot feature importance
plot_importance(model)
pyplot.show()
```
